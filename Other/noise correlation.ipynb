# Ambient noise correlation

```python
import numpy as np
import matplotlib.pyplot as plt

t_min = -200.0
t_max = 200.0
dt = 0.1
n = 1000

dx = 40.0


t = np.arange(1.5*t_min,1.5*t_max,dt)

w1 = t * np.exp(-3.0*t**2)
w2 = (t-dx)*np.exp(-3.0*(t-dx)**2)

plt.plot(t,w1,'k')
plt.xlim((-10.0,10.0))
plt.show()

s = 1.5 * (t_max - t_min) * np.random.rand(n) + 1.5 * t_min 
a = 2.0 * np.random.rand(n) - 1.0

for k in range(n):
    w1 += (t-s[k]) * a[k] * np.exp(-3.0*(t-s[k])**2)
    w2 += ((t-dx)-s[k]) * a[k] * np.exp(-3.0*((t-dx)-s[k])**2)
    
plt.plot(t,w1,'k')
plt.xlim((t_min,t_max))
plt.show()  

plt.plot(t,w2,'k')
plt.xlim((t_min,t_max))
plt.show() 
```

```python
plt.plot(t[500:5500],w1[500:5500],'k')
plt.show()  

tau=300

plt.plot(t[(500+tau):(5500+tau)],w2[(500+tau):(5500+tau)],'k')
plt.show() 

c = w1[500:5500] * w2[(500+tau):(5500+tau)]

plt.plot(t[500:5500],c,'k')
plt.show()
```

```python
tau_max=500

c = np.zeros(tau_max)

for tau in np.arange(0,tau_max):
    c[tau] = np.sum( w1[500:5500] * w2[(500+tau):(5500+tau)] )
    
plt.plot(np.arange(0.0,50.0,dt),c,'k')
plt.show()
```
